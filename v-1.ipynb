**Import Data**

```
from docx import Document
import docx2txt
import pandas as pd 
import string
```

```
Text = docx2txt.process('DataScience.docx')
```

```
type(Text)
```

```
Text= Text.translate(str.maketrans('', '', string.punctuation))
```

```
words = Text.split()
```

```
Words_count = len(words)
```

```
Words_count
```

```
Data_F=[]
while Words_count != 0:
  ipt=[Words_count, words[Words_count-1]]
  Data_F.append(ipt)
  Words_count= Words_count-1
```

```
Data_F
```

```
Table= pd.DataFrame(Data_F,columns=['Position','Word'])
```

```
Table
```

```
Table = Table['Position'].groupby([Table.Word]).apply(list).reset_index()
```
